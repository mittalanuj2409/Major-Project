```
import numpy as np
import pandas as pd
```

```
from numpy import mean
from numpy import std
from sklearn.datasets import make_regression
from sklearn.model_selection import RepeatedKFold
from keras.models import Sequential
from keras.layers import Dense, Dropout

import csv
```

```
#temp_data = np.loadtxt(open('TEMP_DIST.csv', 'r'), delimiter=',', skiprows=1)
#results = np.loadtxt(open('32Sample_mishra.csv', 'r'), delimiter=',', skiprows=1)
data = np.loadtxt(open('2D_Surface_data.csv', 'r'), delimiter=',', skiprows=1)
results = np.loadtxt(open('2D_Surface_result.csv', 'r'), delimiter=',', skiprows=1)
```

```
print(results.shape, data.shape)
```

```
n_inputs, n_outputs = data.shape[1], results.shape[1]
model = Sequential()
```

```
#def get_model(n_inputs, n_outputs):

model.add(Dense(16, input_dim=n_inputs, kernel_initializer='he_uniform', activation='relu'))
#model.add(Dropout(p=0.1))
  # Adding the second hidden layer
model.add(Dense(16, kernel_initializer='uniform', activation='relu'))
# Adding dropout to prevent overfitting
#model.add(Dropout(p=0.1))
model.add(Dense(n_outputs, kernel_initializer='uniform', activation='sigmoid'))
model.compile(loss='mae', optimizer='adam',metrics=["accuracy"])
#	return model

```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(data,results, test_size=0.1,random_state=0)
```

```
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
model.fit(X_train, y_train, batch_size=100, epochs=150)
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
```

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print("Our accuracy is {}%".format(((cm[0][0] + cm[1][1])/57)*100))
```
